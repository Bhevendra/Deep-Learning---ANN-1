```python
# Importing the libararies
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv('aps_data.csv')
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset = dataset.replace("na",np.nan)
```

```python
dataset.isnull().sum()
```

```python
dataset.isnull().sum().sum()
```

```python
pd.options.display.max_rows=1000
```

```python
dataset.isnull().sum()
```

```python
pd.DataFrame(dataset.isnull().sum().sort_values(ascending=False)).rename(columns={0:"Null Value Count"})
```

```python
### filling null values in feature with median of that feature
for feature in [feature for feature in dataset.columns if feature not in ['class']]: dataset[feature]=dataset[feature].fillna(dataset[feature].median())
```

```python
dataset.isnull().sum().sum()
```

```python
dataset.head()
```

```python
X = dataset.drop("class",axis=1)
```

```python
y = dataset["class"]
```

```python
X.info(verbose=True,show_counts=True)
```

```python
pd.get_dummies(y,drop_first=True)
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit_transform(X_train)
sc.transform(X_test)
```
